# SVM Predictive Maintenance Model

This notebook is isolated under `svm/` so the SVM work stays separate from the shared project notebooks and results.

```python
import json
import sys
from pathlib import Path

current_path = Path.cwd().resolve()
repo_root = current_path
for path in [current_path, *current_path.parents]:
    if (path / "data" / "raw" / "predictive_maintenance.csv").exists() and (path / "svm" / "train_svm.py").exists():
        repo_root = path
        break
else:
    raise FileNotFoundError("Could not find the project root from this notebook location.")

sys.path.insert(0, str(repo_root / "svm"))

print("Python executable:", sys.executable)
print("Project root:", repo_root)

try:
    from train_svm import DEFAULT_DATA_PATH, DEFAULT_RESULTS_DIR, train_svm
except ImportError as exc:
    raise RuntimeError(
        "Could not import the SVM dependencies. In Jupyter, select the project "
        "virtualenv kernel/interpreter: .venv/bin/python"
    ) from exc

DEFAULT_DATA_PATH, DEFAULT_RESULTS_DIR
```

```python
metrics = train_svm(
    data_path=DEFAULT_DATA_PATH,
    results_dir=DEFAULT_RESULTS_DIR,
    test_size=0.2,
    random_state=42,
)

metrics["run_id"]
```

```python
summary = {
    "accuracy": metrics["accuracy"],
    "precision": metrics["precision"],
    "recall": metrics["recall"],
    "roc_auc": metrics["roc_auc"],
    "confusion_matrix": metrics["confusion_matrix"],
}

print(json.dumps(summary, indent=2))
```
